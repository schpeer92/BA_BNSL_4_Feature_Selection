### How does GOBNILP works?
source: https://www.jair.org/index.php/jair/article/view/11041/26213 \\

Score based function : $c(B) = \sum_{i\in V} c_i \leftarrow Pa(i,B)$ for (V,B) where B is a set of edges of a possible instance


Algorithm Steps:
1. Solve LP with Convexity Constraints  
2. Find violating cluster constraints (separation problem) 
3. Add Cutting Planes (Branch and Cut) to find integer Solution
