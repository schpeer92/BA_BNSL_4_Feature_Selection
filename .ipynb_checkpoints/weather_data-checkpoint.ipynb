## 1.2 impute missing values or drop row

Simple imputation in a naive way to just work with different time series models and see if Bayesian networks can have an impact.\
First ffill to not implement data leakage and for all still na values we use average imputation


by using mean  we can impute all numerical values and  by mode we can impute categorical 

```python
df_humidity_filled = df_humidity.ffill().fillna(df_humidity.mean())
df_pressure_filled =  df_pressure.ffill().fillna(df_pressure.mean())
df_temperature_filled =  df_temperature.ffill().fillna(df_temperature.mean())
df_weather_description_filled = df_weather_description.ffill().fillna(df_weather_description.mode())
df_wind_direction_filled = df_wind_direction.ffill().fillna(df_wind_direction.mean())
df_wind_speed_filled =  df_wind_speed.ffill().fillna(df_wind_speed.mean())
```

```python
def aggregate_to_daily_vals(df):
    df["datetime"] = pd.to_datetime(df["datetime"])
    df = df.set_index("datetime")
    return df.resample("D").mean().round(2) 
```

```python
df_humidity_agg = aggregate_to_daily_vals(df_humidity_filled)
df_pressure_agg =  aggregate_to_daily_vals(df_pressure_filled)
df_temperature_agg =  aggregate_to_daily_vals(df_temperature_filled) 
df_wind_direction_agg = aggregate_to_daily_vals(df_wind_direction_filled)
df_wind_speed_agg =  aggregate_to_daily_vals(df_wind_speed_filled)
```

# 3. Time Series Modeling 

```python
df_weather_data = df_weather_data.set_index("datetime") # in the following cells this is needed for time indexing
```

```python
import statsmodels.api as sm

class VAR:
    
    def __init__(self,data_train:pd.DataFrame,start_date:str, end_date:str,p:int,mask:np.array=None,)->None:
        """
        Args:
            data_train (pd.DataFrame): Time Series dataset with datetime index
            start_date (str): start date of training period
            end_date (str): end date of training period
            p (int): number of time lag variables
            mask (np.array, optional): masking of variables to learn linear regression. default none, will take all variables
        """
        assert len(data_train[data_train.index <start_date]) >= p , "number of records before start_date is to small. Time shifts will produces nan values!" 

        self.data_train = data_train[:end_date]
        
        if mask is None:
            self.mask = np.ones((self.data_train.shape[1],self.data_train.shape[1]))
        else:
            self.mask = mask
        self.p = p
        self.models = {}
        self.start_date = start_date
        self.end_date = end_date

    def insert_shifted_data(self, data:pd.DataFrame)-> pd.DataFrame:
        """Takes Dataframe and adds time lag features for all columns

        Args:
            data (pd.DataFrame): Dataframe of which we build time lag features

        Returns:
            pd.DataFrame: DataFrame with additional time lag features 
        """
        cols_to_shift = data.columns
        for shift in range(1,self.p+1):
            for col in cols_to_shift:
                data[f"{col}_{shift}_shifted"] = data[col].shift(shift)
        
        return data

    def train_model_per_col(self)->None:
        """Runs for each column multivariate linear regression by only taking into account features that are unmasked
        """
        for i,col in enumerate(self.data_train.columns ):
            mask = np.squeeze(np.asarray(self.mask[:,i])>0)
            data = self.data_train.iloc[:,mask].copy()
            #
            data_shifted = self.insert_shifted_data(data)
            data_shifted = data_shifted[[col for col in data_shifted.columns if col.endswith("_shifted")]]
            data_shifted = data_shifted[self.start_date:]
            col_model = sm.OLS(self.data_train[self.start_date:][col].values, data_shifted.values).fit()
            self.models[col] = col_model

    
    def predict_per_col(self, data:pd.DataFrame)-> pd.DataFrame:
        """Run prediction on DataFrame 

        Args:
            data (pd.DataFrame): DataFrame to predict, has to have at least p rows and datetime index

        Returns:
            pd.DataFrame: dataframe with predictions and actual values
        """
        data_pred = data.copy()
        for i, col in enumerate(self.data_train.columns):
            data_pred[f"{col}_pred"] = None
            mask = np.squeeze(np.asarray(self.mask[:,i])>0)
            data_shifted = self.insert_shifted_data(data.iloc[:,mask])
            data_shifted = data_shifted.fillna(0)
            data_shifted = data_shifted[[col for col in data_shifted.columns if col.endswith("_shifted")]]
            data_pred[f"{col}_pred"] = self.models[col].predict(data_shifted.values)
        return data_pred


    def recursive_prediction(self,prediction_steps:int = 7)->pd.DataFrame:
        """runs prediction_steps many iterations of recursive forecasting which uses iteratively prediction values as input

        Args:
            prediction_steps (int, optional):number of prediction steps Defaults to 7.

        Returns:
            pd.DataFrame: Dataframe with predictions as columns for prediction_steps many values
        """
        data = self.data_train[self.start_date:self.end_date]
        start = f"{str(pd.to_datetime(self.end_date).year +1)}-01-01"
        for prediction_step in pd.date_range(start= start, periods=prediction_steps, freq="D"):
            data_pred = data.tail(self.p).append(
                pd.DataFrame.from_dict({"datetime":[prediction_step], "humidity":[0], "pressure":[0],"wind_direction":[0], "temperature":[0],"wind_speed":[0]}).set_index("datetime"))
            data_pred = self.predict_per_col(data_pred)
            data_cols = data.columns
            data_pred.drop(columns=data_cols, inplace=True)
            data_pred.columns = data_cols
            data = data.append(data_pred.tail(1))
        
        return data[start:]
            
            


    def evaluate_prediction(self,Y,Y_pred):
        pass
            
```

```python
def find_best_p(start_date:str="2012-11-01",end_date:str="2015-12-31",mask:np.array=None)->dict:
    measurements = {}
    for city in ["Vancouver", "Charlotte","Haifa", "San Diego"]:#df_weather_data["city"].unique():
        measurements[city] = {}
        for p in range(1,8): #at most one week
            var_model = VAR(df_weather_data[lambda x: x["city"]==city].drop(columns = "city"),start_date=start_date,end_date=end_date,p=p, mask=mask)
            var_model.train_model_per_col()
            df_pred = var_model.predict_per_col(df_weather_data["2016-01-01":][lambda x: x["city"]==city].drop(columns="city"))
            data_cols = df_weather_data["2016-01-01":][lambda x: x["city"]==city].drop(columns="city").columns
            df_pred.drop(columns=data_cols, inplace=True)
            df_pred.columns = data_cols
            rmse = (((df_pred - df_weather_data["2016-01-01":][lambda x: x["city"]==city].drop(columns="city"))**2)**0.5).mean()
            measurements[city][p] = rmse

    return measurements
```

```python
def find_best_p(start_date:str="2012-11-01",end_date:str="2015-12-31",mask:np.array=None)->dict:
    measurements = {}
    for city in ["Vancouver", "Charlotte","Haifa", "San Diego"]:#df_weather_data["city"].unique():
        measurements[city] = {}
        for p in range(1,8): #at most one week
            var_model = VAR(df_weather_data[lambda x: x["city"]==city].drop(columns = "city"),start_date=start_date,end_date=end_date,p=p, mask=mask)
            var_model.train_model_per_col()
            df_pred = var_model.predict_per_col(df_weather_data["2016-01-01":][lambda x: x["city"]==city].drop(columns="city"))
            data_cols = df_weather_data["2016-01-01":][lambda x: x["city"]==city].drop(columns="city").columns
            df_pred.drop(columns=data_cols, inplace=True)
            df_pred.columns = data_cols
            rmse = (((df_pred - df_weather_data["2016-01-01":][lambda x: x["city"]==city].drop(columns="city"))**2)**0.5).mean()
            measurements[city][p] = rmse

    return measurements
```

```python
def evalulate_for_fixed_p(start_date:str="2012-11-01",end_date:str="2015-12-31", p = 3):
    cities = ["Vancouver", "Charlotte","Haifa", "San Diego"]
    measurements = {}
    data_cols = df_weather_data["2016-01-01":][lambda x: x["city"]=="Vancouver"].drop(columns="city").columns
    for city in cities:
        measurements[city] = {}
        for i in range(2):
            mask = None if i==0 else bn_adj_mask
            var_model =  VAR(df_weather_data[lambda x: x["city"]==city].drop(columns = "city"),start_date=start_date,end_date=end_date,p=p, mask=mask)
            var_model.train_model_per_col()
            df_pred = var_model.predict_per_col(df_weather_data["2016-01-01":][lambda x: x["city"]==city].drop(columns="city"))
            df_pred.drop(columns=data_cols, inplace=True)
            df_pred.columns = data_cols
            rmse = (((df_pred - df_weather_data["2016-01-01":][lambda x: x["city"]==city].drop(columns="city"))**2)**0.5).mean()
            measurements[city][i] = rmse

    return measurements
```
