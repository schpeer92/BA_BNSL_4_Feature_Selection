## Prepare Airquality UCI
https://archive.ics.uci.edu/ml/datasets/Air+Quality

```python
df_air_quality = pd.read_csv("../data/01_raw/AirQualityUCI.csv", sep = ";")
```

```python
df_air_quality.head()
```

```python
df_air_quality.tail(
)
```

```python
"string".startswith("unnamed")
```

```python
[col for col in df_air_quality.columns if col.startswith("Unnamed") == False]
```

Delete empty cols

```python
df_air_quality = df_air_quality[[col for col in df_air_quality.columns if col.startswith("Unnamed") == False]]
```

Replace comma in numerics with dot to cast every columns as float

```python
cols_air_quality_numeric = [col for col in df_air_quality.columns if col not in {"Date","Time"}]
```

```python
for col in cols_air_quality_numeric:
    try:
        df_air_quality[col] = df_air_quality[col].str.replace(",",".").astype("float")
    except:
        next
```

```python
df_air_quality.dtypes
```

Dataset contains empty values "-200" => ffill since first value is non empty => no data leakage

```python
df_air_quality = df_air_quality[lambda x: x["Date"].notnull()]
```

```python
df_air_quality = df_air_quality.replace(-200,method = "ffill")
```

```python
df_air_quality.to_csv("../data/02_preprocessed/air_preprocessed.csv", index = False)
```

## Prepare Metro_Interstate_Traffic_Volume
